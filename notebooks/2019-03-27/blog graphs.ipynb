# Blog Graphs

## Rate of Climb 75 Knots

Best rate of climb tests of BD-4C N2468Z on 27 March 2019 at St. Charles County Smartt Airport (KSET).

## Setup

```python
%matplotlib inline

from MglEfisPlotter import *
```

Load the flights from the file exported from the MGL iEFIS.

```python
datafile = 'IEFISBB.DAT'

flights = createFlights(datafile)
for i in range(0, len(flights)):
    print(i, ':', flights[i])

f = flights[6]
p = Plot(f)
print(f)
```

## Graphs

```python
p.plot2(['asi', 'vsi', 'densityAltitude'], ['Airspeed', 'Vertical Speed', 'Altitude']).save('climb-overview.png')
```

```python
p.plot2(['asi', 'vsi', 'densityAltitude'], ['Airspeed', 'Vertical Speed', 'Altitude'], xlim=(6, 10))
p.save('climb-test.png')
```

```python
p.plot2(['asi', 'rpm', 'manifoldPressure'], ['Airspeed', 'Engine RPM', 'Manifold Pressure'], xlim=(6, 10))
p.save('climb-test-power.png')
```

```python
p.plot2(['asi', 'cht', 'oilTemperature1', 'oilPressure1'], 
        ['Airspeed', 'Cylinder Head Temperature', 'Oil Temperature', 'Oil Pressure'], 
        xlim=(5, 15))
p.save('climb-test-engine.png')
```
