# Data by Airspeed, Summarized for All Flights

```python
from MglEfisPlotter import *
```

Load the flights from the file exported from the MGL iEFIS.

```python
datafile = 'IEFISBB.DAT'

flights = createFlights(datafile)
for i in range(0, len(flights)):
    print(i, ':', flights[i])
```

## ASI vs CHT

```python
for i in range(3, 7):
    p = Plot(flights[i])
    p.plot2(['asi', 'cht'])
    p.show()
```

## ASI vs Oil Temperature & Pressure

```python
for i in range(3, 7):
    p = Plot(flights[i])
    p.plot2(['asi', 'oilTemperature1', 'oilPressure1'])
    p.show()
```

## ASI vs Pitch Angle

```python
for i in range(3, 7):
    p = Plot(flights[i])
    p.plot2(['asi', 'pitchAngle'])
    p.show()
```
