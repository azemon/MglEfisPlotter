## Fuel Flow at 75% power 65% power, and 55% power

```python
labels = ['Engine RPM', 'Manifold Pressure', 'Fuel Flow']
p.plot2(['rpm', 'manifoldPressure', 'fuelFlow'], labels)\
 .save('fuel-flow-overview.png')
```

## 75% Power, 30 Minute Cruise

```python
p.plot2(['rpm', 'manifoldPressure', 'fuelFlow'], labels, xlim=(18, 48), ylim=(2300, 2500))\
 .save('fuel-flow-75.png')
p.plot2(['manifoldPressure'], ['Manifold Pressure'], xlim=(18, 48), ylim=(23, 25)).save('manifold-pressure-75')
```
