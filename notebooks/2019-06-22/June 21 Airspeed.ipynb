```python
from mgl_efis_plotter import *
import pandas as pd
```

```python
flights = create_flights('IEFIS.REC', Config)
for i in range(0, len(flights)):
    print(i, flights[i])
```

## Overview

```python
p = Plot(flights[29])
p.plot2(['asi', 'pAltitude', 'tas']).show()
```

```python
p.plot2(['rpm', 'manifoldPressure', 'fuelFlow']).show()
```

## Airspeed

```python
xlim = (25, 85)
p.plot2(['asi', 'pAltitude', 'tas'], xlim=xlim).show()
p.plot2(['asi', 'pAltitude', 'tas'], xlim=(150, 200)).show()
```

```python
asi_data = p.flight.list_data('asi')
asi = pd.DataFrame(asi_data, index=asi_data['minutes'], columns=['asi'])

tas_data = p.flight.list_data('tas')
tas = pd.DataFrame(tas_data, index=tas_data['minutes'], columns=['tas'])

ff_data = p.flight.list_data('fuelFlow')
ff = pd.DataFrame(ff_data, index=ff_data['minutes'], columns=['fuelFlow'])
```

```python
print('Average ASI at 8,500 feet =', float(asi.loc[25:85].mean()), 'knots')
print('Average TAS at 8,500 feet =', float(tas.loc[25:85].mean()), 'knots')
print('Average fuel flow at 8,500 feet =', float(ff.loc[25:85].mean()), 'GPH')
```

```python
print('Average ASI at 10,500 feet =', float(asi.loc[150:200].mean()), 'knots')
print('Average TAS at 10,500 feet =', float(tas.loc[150:200].mean()), 'knots')
print('Average fuel flow at 10,500 feet =', float(ff.loc[150:200].mean()), 'GPH')
```
