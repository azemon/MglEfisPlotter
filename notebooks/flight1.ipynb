# First Flight Data N2468Z

Data from the first flight of BD-4C N2468Z on 13 February 2019 at St. Charles County Smartt Airport (KSET).

## Setup

```python
%matplotlib inline

import sys
sys.path.append('..')
from MglEfisPlotter import *
```

Load the flights from the file exported from the MGL iEFIS.

```python
datafile = '../data/IEFISBB.DAT'
minTimestamp = 429600874
maxTimestamp = 1000000000

flights = createFlights(datafile, minTimestamp, maxTimestamp)
```

Select the last flight in the datafile, which is the first test flight. The others were taxi tests.

```python
p = Plot(flights[-1])
print(p.flight)
```
