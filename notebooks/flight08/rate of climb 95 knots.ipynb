## Rate of Climb 95 Knots

Best rate of climb tests of BD-4C N2468Z on 27 March 2019 at St. Charles County Smartt Airport (KSET).

## Setup

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.append('../..')
from MglEfisPlotter import *
```

```python
plt.rcParams['figure.figsize'] = (14, 10)
```

Load the flights from the file exported from the MGL iEFIS.

```python
datafile = 'IEFISBB.DAT'

flights = createFlights(datafile)
for i in range(0, len(flights)):
    print(i, ':', flights[i])

f = flights[3]
p = Plot(f)
print(f)
```

## Graphs

```python
p.plot2(['asi', 'vsi', 'pAltitude'])
p.show()
```

```python
t = f.listData('asi')
asi = pd.DataFrame(t, columns=['minutes', 'asi'])

t = f.listData('vsi')
vsi = pd.DataFrame(t, columns=['minutes', 'vsi'])

t = f.listData('pAltitude')
pAltitude = pd.DataFrame(t, columns=['minutes', 'pAltitude'])

xlim=(3, 6)

merged = asi.merge(vsi)
merged = merged.merge(pAltitude)

merged.plot(x='minutes', secondary_y=['vsi', 'pAltitude'], xlim=xlim)
```
