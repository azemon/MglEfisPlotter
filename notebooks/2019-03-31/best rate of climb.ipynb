```python
p = Plot(flights[0])
xlim=(3.25, 5.25)
p.plot2(['asi', 'vsi', 'pitchAngle'], xlim=xlim, ylim=(75, 130)).show()
p.flight.save_csv('rateofclimb80knots.csv', ['asi', 'vsi', 'pitchAngle', 'densityAltitude'])
```
